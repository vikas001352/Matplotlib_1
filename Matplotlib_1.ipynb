```python
ANS-1


Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in 2D and 3D. It provides a wide range of high-quality plots and charts, making it an essential tool for data visualization in various fields.

Key features and uses of Matplotlib:

Data Visualization: Matplotlib allows users to visualize and explore data in a graphical format, helping to gain insights, identify patterns, and communicate findings effectively.

Publication-Quality Plots: It offers the flexibility to create publication-quality plots suitable for scientific publications, reports, presentations, and data analysis.

Interactive Plots: Matplotlib can be integrated with interactive tools and frameworks, enabling users to create interactive visualizations for data exploration.

Customizability: The library provides extensive customization options, allowing users to fine-tune plots and charts to meet specific requirements.

2D and 3D Plotting: Matplotlib supports both 2D and 3D plotting, enabling the creation of a wide range of visualizations.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line Plot: A simple plot to visualize the relationship between two variables by connecting data points with lines.

Scatter Plot: Used to display the distribution of data points in a 2D space, where each point represents an observation.

Bar Plot: Suitable for comparing categorical data by displaying rectangular bars of different heights or lengths.

Histogram: Used to visualize the frequency distribution of continuous or discrete data, dividing it into bins.

Pie Chart: Represents data as slices of a circular pie, where each slice represents a portion of the whole.

These are just a few examples, and Matplotlib offers many other plot types and functionalities to cater to diverse visualization needs.





ANS-2


import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create the scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()




ANS-3


import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Data for line plots
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid for subplots and plot the lines
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout and display the subplots
plt.tight_layout()
plt.show()



ANS-4



import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 5))  # Optional: Set the figure size
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')

# Display the bar plot
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 5))  # Optional: Set the figure size
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')

# Display the horizontal bar plot
plt.show()




ANS-5



Box plots are used for the following purposes:

Visualizing Data Spread: Box plots allow us to understand the spread and variability of the dataset by identifying quartiles and outliers.

Identifying Skewness: They help in identifying if the data is symmetrically distributed or skewed to the left or right.

Comparing Distributions: Box plots are useful for comparing the distribution of multiple datasets side by side.

Detecting Outliers: They highlight potential outliers, which are data points that fall significantly far from the bulk of the data.

Now, let's create a box plot using the given data



import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list for plotting
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 5))  # Optional: Set the figure size
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Values')

# Display the box plot
plt.show()


```
